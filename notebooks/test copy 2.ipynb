```python
from dotenv import load_dotenv

load_dotenv()  

from openai_agent import orchestrator, email_agent, summarizer_agent, translator_agent

print("Agents loaded successfully.")
print(f"Orchestrator handoffs: {[a.name for a in orchestrator.handoffs]}")
print(f"Email handoffs: {[a.name for a in email_agent.handoffs]}")
print(f"Summarizer handoffs: {[a.name for a in summarizer_agent.handoffs]}")
print(f"Translator handoffs: {[a.name for a in translator_agent.handoffs]}")
```

```python
print(email_agent.handoffs)
print(summarizer_agent.handoffs)
print(translator_agent.handoffs)
```

---
### Test 1 — Orchestrator routes a summarisation request

```python
orchestrator_result = await orchestrator.run_agent(
    "Summarise the following: "
    "Artificial intelligence is transforming every industry. "
    "From healthcare to finance, AI systems are automating repetitive tasks, "
    "improving decision-making, and unlocking insights from massive datasets."
)

print("Final output:", orchestrator_result.final_output)
print("Last agent:", orchestrator_result.last_agent.name)
```

---
### Test 2 — Orchestrator routes a translation request

```python
orchestrator_result_translator = await orchestrator.run_agent("Translate 'Good morning, how are you?' into French.")

print("Final output:", orchestrator_result_translator.final_output)
print("Last agent:", orchestrator_result_translator.last_agent.name)
```

```python
examples = [
    "Draft an email about the Q3 product roadmap update.",
    "Summarise the following: Artificial intelligence is transforming industries "
    "at an unprecedented pace. From healthcare to finance, AI-driven tools are "
    "automating repetitive tasks, uncovering insights from massive datasets, and "
    "enabling personalised experiences at scale.",
    "Translate 'Good morning, how are you?' into French.",
]

for user_input in examples:
    print("\n" + "=" * 60)
    print(f"INPUT: {user_input}\n")
    result = await orchestrator.run_agent(user_input)
    print(f"OUTPUT TYPE: {type(result).__name__}")
    print(f"OUTPUT: {result.final_output}")
```
